#**MatPlotlibで自由なフォントを使用する**
準備： gdrive内にフォントのフォルダを作っておき、その中にttfファイルをダウンロードしておく

folder.png

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import glob

from google.colab import drive
drive.mount("/content/gdrive")

print("")
print(glob.glob("/content/gdrive/MyDrive/Font/*"))
```

```python
import matplotlib.font_manager as fm

FONT_PATH = '/content/gdrive/MyDrive/Font/arial.ttf'
fp = fm.FontProperties(fname=FONT_PATH)

print(fp.get_name())
print(fp.get_family())

plt.rcParams["font.family"] = fp.get_family()


plt.rcParams["font.size"] = 30
plt.title('Raspberry Pi', fontproperties=fp)
plt.ylabel('aaa', fontproperties=fp)


#解像度を上げてセーブする
plt.savefig("/content/default_dpi.png", bbox_inches="tight", pad_inches=0.05, format="png")
plt.savefig("/content/300_dpi.png", bbox_inches="tight", pad_inches=0.05, format="png", dpi=300)
```
